```python
%matplotlib notebook

from matplotlib.animation import FuncAnimation
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_polygon(n):

    # Calculate the vertices of the polygon
    angle = 2 * np.pi / n
    verticesx = [np.cos(i * angle) for i in range(n)]
    verticesy = [np.sin(j * angle) for j in range(n)]
    return np.append(verticesx,verticesx[0]),np.append(verticesy,verticesy[0])

def repeater(arr):
    re = []
    for i in range(len(arr)):
        re = np.append(re,arr[i])
        re = np.append(re,arr[i])
        
    return re[:-1]
  
    
def splitter(arr):
    sp = []
    for i in range(len(arr)-2):
        sp = np.append(sp,arr[i])
        sp = np.append(sp,arr[i+1])
        
    sp = np.append(sp,arr[-1])
    
    return sp
```

```python
xs,ys = create_polygon(4)
xt,yt = create_polygon(3)
xp,yp = create_polygon(5)
xh,yh = create_polygon(6)
xse,yse = create_polygon(7)
xo,yo = create_polygon(8)
```

```python
def morph(x1, y1, x2, y2, alpha): 
    xm = alpha * x1 + (1-alpha) * x2 
    ym = alpha * y1 + (1-alpha) * y2
    return xm, ym
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [],'r')


def init(): 
    ax.set_xlim([0, 1]) 
    ax.set_ylim([0,1]) 
    return ln,
def update(frame):
    if 0 <= frame < 0.2:
        # Morph from triangle to square
        xtt,ytt = repeater(xt),repeater(yt)
        xss,yss = splitter(xs),splitter(ys)
        
        t = (frame)*5
        xdata, ydata = morph(xss, yss, xtt, ytt, t)
    elif 0.2 <= frame < 0.4:
        # Morph from square to pentagon
        xss,yss = repeater(xs),repeater(ys)
        xpp,ypp = splitter(xp),splitter(yp)
        
        t = (frame-0.2)*5
        xdata, ydata = morph(xpp, ypp, xss, yss, t)
    elif 0.4 <= frame < 0.6:
        # Morph from pentagon to hexagon
        xpp,ypp = repeater(xp),repeater(yp)
        xhh,yhh = splitter(xh),splitter(yh)
        
        t = (frame-0.4)*5
        xdata, ydata = morph(xhh, yhh, xpp, ypp, t)
    elif 0.6 <= frame < 0.8:
        # Morph from hexagon to septagon
        xhh,yhh = repeater(xh),repeater(yh)
        xsese,ysese = splitter(xse),splitter(yse)
        
        t = (frame-0.6)*5
        xdata, ydata = morph(xsese, ysese, xhh, yhh, t)
    elif 0.8 <= frame < 1.0:
        # Morph from septagon to octagon
        xsese,ysese = repeater(xse),repeater(yse)
        xoo,yoo = splitter(xo),splitter(yo)
        
        t = (frame-0.8)*5
        xdata, ydata = morph(xoo, yoo, xsese, ysese, t)
    elif 1.0 <= frame < 1.2:
        # Morph from octagon to septagon
        xsese,ysese = repeater(xse),repeater(yse)
        xoo,yoo = splitter(xo),splitter(yo)
        
        t = (frame-1.0)*5
        xdata, ydata = morph(xsese, ysese, xoo, yoo, t)
    elif 1.2 <= frame < 1.4:
        # Morph from septagon to hexagon
        xhh,yhh = repeater(xh),repeater(yh)
        xsese,ysese = splitter(xse),splitter(yse)
        
        t = (frame-1.2)*5
        xdata, ydata = morph(xhh, yhh, xsese, ysese, t)
    elif 1.4 <= frame < 1.6:
        # Morph from hexagon to pentagon
        xpp,ypp = repeater(xp),repeater(yp)
        xhh,yhh = splitter(xh),splitter(yh)
        
        t = (frame-1.4)*5
        xdata, ydata = morph(xpp, ypp, xhh, yhh, t)
    elif 1.6 <= frame < 1.8:
        # Morph from pentagon to square
        xss,yss = repeater(xs),repeater(ys)
        xpp,ypp = splitter(xp),splitter(yp)
        
        t = (frame-1.6)*5
        xdata, ydata = morph(xss, yss, xpp, ypp, t)
    else:
        # Morph from square to triangle
        xtt,ytt = repeater(xt),repeater(yt)
        xss,yss = splitter(xs),splitter(ys)
        t = (frame-1.8)*5
        xdata, ydata = morph(xtt, ytt, xss, yss, t)
        
    ln.set_data(xdata, ydata)
    return ln

#     if frame<0.5:
#         frame = frame*2
#         xtt = repeater(xt)
#         ytt = repeater(yt)
        
#         xss = splitter(xs)
#         yss = splitter(ys)
        
#         xdata,ydata = morph(xss,yss,xtt,ytt,frame)
#     else:
#         frame = (frame-0.5)*2
#         xss = repeater(xs)
#         yss = repeater(ys)
        
#         xpp = splitter(xp)
#         ypp = splitter(yp)
        
#         xdata,ydata = morph(xpp,ypp,xss,yss,frame)
        
#     ln.set_data(xdata, ydata)
#     return ln,

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])

anim = FuncAnimation(fig, update, frames=np.linspace(0, 2, 500), blit=True,interval=10,repeat=True)
plt.show()
```
