```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
def create_polygon(n):

    # Calculate the vertices of the polygon
    angle = 2 * np.pi / n
    verticesx = [np.cos(i * angle) for i in range(n)]
    verticesy = [np.sin(j * angle) for j in range(n)]
    return np.append(verticesx,verticesx[0]),np.append(verticesy,verticesy[0])

xs,ys = create_polygon(4)
xt,yt = create_polygon(3)
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'r')

def init(): 
    ax.set_xlim(-1.2, 1.2) 
    ax.set_ylim(-1.2, 1.2) 
    return ln,

def morph(xs,ys,xt,yt,t):
    xm1 = xs*t+xt*(1-t)
    ym1 = ys*t+yt*(1-t)
    
    xm2 = xt*t+xs*(1-t)
    ym2 = yt*t+ys*(1-t)
    
    x = xm2*t + xm1*(1-t)
    y = ym2*t + ym1*(1-t)
    
    return x,y   

def update(frame):
    # xdata.append(frame)
    # ydata.append(np.sin(frame))
    xdata, ydata = morph(xs, ys, xt, yt, frame) 
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 1, 120), init_func=init, blit=True, interval=10, repeat=True)
plt.show()
```
