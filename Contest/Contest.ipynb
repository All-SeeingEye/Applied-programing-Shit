```python
import numpy as np

def split_array_min_distance(arr):
    sorted_arr = np.sort(arr)
    cum_sum = np.cumsum(sorted_arr)
    total_sum = np.sum(sorted_arr)
    min_diff = np.inf
    split_point = None
    
    for i in range(1, len(sorted_arr)):
        diff = abs(cum_sum[i-1] - (total_sum - cum_sum[i-1]))
        if diff < min_diff:
            min_diff = diff
            split_point = i
    
    return sorted_arr[:split_point], sorted_arr[split_point:]

```
